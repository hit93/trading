```python
import pandas as pd
df = pd.read_csv('bnf.csv')
df.head(50)
```

```python
ohlc_dict = {'open':'first', 'high':'max', 'low':'min', 'close': 'last'}
```

```python
df.index = pd.to_datetime(df['timestamp'])
```

```python
df.index = df.index.map(lambda x: x.replace(second=0))
```

```python
df = df.resample('3Min').apply(ohlc_dict)
df.dropna(inplace = True)
# pd.set_option('display.max_rows', None)
df.head(50)
```

```python
df = df.join(df[['open','high','low','close']].ta.supertrend(length = 10,multiplier = 1.5))
df = df.join(df[['open', 'high','low','close']].ta.supertrend(length = 10,multiplier = 1))
```

```python
df = df.join(df.ta.macd(fast = 12, slow = 20))
```

```python
df = df.join(df.ta.ema(length = 200))
df = df.join(df.ta.ema(length = 100))
df = df.join(df.ta.ema(length = 50))
```
