```python
from backtesting.test import GOOG
import pandas as pd
df = pd.read_csv('bnf_oi.csv', index_col = 0)

df.rename(columns = {'open':'Open', 'high':'High', 'low':'Low', 'close':'Close'}, inplace = True)
df.index = pd.to_datetime(df.index)
df
```

```python
pd.set_option('display.max_rows', 100)
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover 
from backtesting.test import SMA, GOOG
```

```python
def macd():
    return df.MACD_12_20_9
def macds():
    return df.MACDs_12_20_9
def close():
    return df.Close
def high():
    return df.High
def low():
    return df.Low
def hour():
    return pd.to_datetime(df.index).hour
def minute():
    return pd.to_datetime(df.index).minute
def ema50():
    return df.EMA_50
def ema100():
    return df.EMA_100
def ema200():
    return df.EMA_200
 
class intraday(Strategy):
    def init(self):
        super().init()
        price = self.data.Close
        self.OI = self.I(SMA, self.data.diff, 15)
        self.macd = self.I(macd)
        self.macds = self.I(macds)
        self.close = self.I(close)
        self.hour = self.I(hour)
        self.minute = self.I(minute)
        self.high = self.I(high)
        self.low = self.I(low)
        self.ema50 = self.I(ema50)
        self.ema100 = self.I(ema100)
        self.ema200 = self.I(ema200)

    def next(self):
        if ( self.macds < 0 and
             crossover(self.macd,self.macds) and
             self.hour > 9 and
             self.hour < 15 and
             self.OI > 1000000 and
             self.ema100 < self.ema50 and
             self.close> self.ema50 and
             self.OI > 0.8*self.OI[-10]):
            self.buy(sl = min(self.low[-5:-1])-10)
        for i in self.trades:
            if i.pl_pct>0.003:
                i.sl = self.low[-2]
        if self.hour == 15 and self.minute==15:
            self.position.close()
        
        

# class SmaCross(Strategy):
#     def init(self):
#         price = self.data.Close
#         self.ma1 = self.I(SMA, price, 10)
#         self.ma2 = self.I(SMA, price, 20)

#     def next(self):
#         if crossover(self.ma1, self.ma2):
#             self.buy()
#         elif crossover(self.ma2, self.ma1):
#             self.sell()

bt = Backtest(df, intraday, cash = 1000000, commission=.0002,
              exclusive_orders=True, margin=1 )
stats = bt.run()
stats
```
